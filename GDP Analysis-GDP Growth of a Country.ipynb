```python
data =df_pr.values 
for i in range (1,len(data)):

    prev =data[i-1][3]
    cur=data[i][3]
    print( data[i][2] , round(((cur-prev)/prev)*100,2))      #[3]-specifying take column 3 data 
```

```python
data = df_pr.values 

gdp_change=[0]

for i in range (1,len(data)):

    prev =data[i-1][3]
    cur = data[i][3]
    gdp_change.append(round(((cur-prev)/prev)*100,2) )     #[3]-specifying take column 3 data 
```

```python
gdp_change
```

```python
df_pr.assign(GDP = gdp_change) #adding new column to data
```
