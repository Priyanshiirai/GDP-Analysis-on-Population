```python
import pandas as pd 
df= pd.read_csv('gdp.csv')
```

```python
df.head()
```

# GDP ANALYSIS FOR INDIA 

```python
df_pr = df[df['Country Name'] == 'India']
```

```python
data = df_pr.values

gdp_change = [0]
for i in range (1, len(data)):
    prev = data[i-1][3]
    cur = data[i][3]
    gdp_change.append(round(((cur-prev)/prev)*100,2))

df_pr = df_pr.assign(GDP = gdp_change)
```

```python
df_pr
```
